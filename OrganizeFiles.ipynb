```python
import os
import shutil

SUMMARYDIR = '/content/drive/MyDrive/AI_Lec_run/summary'
ENGLISHEDDIR = '/content/drive/MyDrive/AI_Lec_run/englished'
TXT_DIR = os.path.join(SUMMARYDIR, 'txt_files')
DOCX_DIR = os.path.join(SUMMARYDIR, 'docx_files')

# 폴더가 없으면 생성
if not os.path.exists(TXT_DIR):
    os.makedirs(TXT_DIR)
if not os.path.exists(DOCX_DIR):
    os.makedirs(DOCX_DIR)

# 파일 분리
for filename in os.listdir(SUMMARYDIR):
    if filename.endswith('.txt'):
        shutil.move(os.path.join(SUMMARYDIR, filename), os.path.join(TXT_DIR, filename))
    elif filename.endswith('.docx'):
        shutil.move(os.path.join(SUMMARYDIR, filename), os.path.join(DOCX_DIR, filename))
```

```python
def categorize_by_subject(directory):
    subject_dirs = {}  # 과목명을 키로, 해당 폴더 경로를 값으로 하는 딕셔너리

    for filename in [i for i in os.listdir(directory) if (i.endswith('.docx') or i.endswith('.txt')) ]:
        subject_name = filename.split('_')[0]  # 과목명 추출
        subject_folder = os.path.join(directory, subject_name)  # 과목 폴더 경로 생성

        # 과목 폴더가 없다면 생성
        if subject_name not in subject_dirs:
            if not os.path.exists(subject_folder):
                os.mkdir(subject_folder)
            subject_dirs[subject_name] = subject_folder

        # 파일을 해당 과목 폴더로 이동
        shutil.move(os.path.join(directory, filename), os.path.join(subject_dirs[subject_name], filename))

# 함수 사용 예시

categorize_by_subject('/content/drive/MyDrive/AI_Lec_run/summary/docx_files')

```
